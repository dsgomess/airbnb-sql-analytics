```python
import pandas as pd 
import csv

```

Tratando a planilha listings

```python
df = pd.read_csv(
    "data/listings.csv", #aqui voce deve colocar o caminho do arquivo que voce baixou
    sep=",",
    quotechar='"',
    encoding="utf-8",
    engine="python"
)

# remover quebras de linha internas
df = df.replace(r'\r\n|\r|\n', ' ', regex=True)

# remover vírgula do price (separador de milhar)
df["price"] = df["price"].replace('[\$,]', '', regex=True)

colunas_desejadas = [
    "id",
    "host_id",
    "neighbourhood_cleansed",
    "latitude",
    "longitude",
    "property_type",
    "room_type",
    "accommodates",
    "bathrooms",
    "bedrooms",
    "beds",
    "price",
    "minimum_nights",
    "maximum_nights",
    "number_of_reviews",
    "review_scores_rating",
    "availability_365"
]

df = df[colunas_desejadas]

df.to_csv(
    "data/listings_filtered.csv",
    index=False,
    encoding="utf-8",
    sep=",",
    quoting=csv.QUOTE_NONE,
    escapechar="\\"
)
print("Arquivo tratado corretamente.")
```
